```python
students = {}

# Input student names and scores
num_students = int(input("Enter number of students: "))

for i in range(num_students):
    name = input(f"Enter student {i + 1} name: ")
    score = float(input(f"Enter {name}'s score: "))
    students[name] = score

# Display all student scores
print("\nStudent Scores:")
for name, score in students.items():
    print(f"{name}: {score}")

# Calculate class average
average_score = sum(students.values()) / len(students)
print(f"\nClass Average: {average_score:.2f}")

# Find the top scorer
top_student = max(students, key=students.get)
top_score = students[top_student]

print(f"Top Scorer: {top_student}")
print(f"Highest Score: {top_score}")
```
